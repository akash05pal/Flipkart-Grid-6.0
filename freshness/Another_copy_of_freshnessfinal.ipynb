```python
import torch
import cv2
import time
from pathlib import Path
import numpy as np
from models.experimental import attempt_load
from utils.general import non_max_suppression  # Only import non_max_suppression
from utils.torch_utils import select_device

# Define the letterbox function
def letterbox(img, new_shape=(640, 640), color=(114, 114, 114), auto=True, scaleFill=False, scaleup=True):
    shape = img.shape[:2]  # current shape [height, width]
    if isinstance(new_shape, int):
        new_shape = (new_shape, new_shape)  # if single int provided, make it a tuple
    ratio = min(new_shape[0] / shape[0], new_shape[1] / shape[1])  # ratio for resizing
    new_unpad = int(round(shape[1] * ratio)), int(round(shape[0] * ratio))  # new shape
    dw, dh = new_shape[1] - new_unpad[0], new_shape[0] - new_unpad[1]  # width and height padding
    if auto:  # resize and pad
        dw, dh = dw % 2, dh % 2  # make it even
    elif scaleFill:  # scale fill
        dw, dh = 0, 0
        new_unpad = new_shape[1], new_shape[0]
        ratio = new_shape[1] / shape[1]  # update ratio

    top, bottom = dh // 2, dh - (dh // 2)  # padding
    left, right = dw // 2, dw - (dw // 2)  # padding

    img = cv2.resize(img, new_unpad, interpolation=cv2.INTER_LINEAR)  # resize
    img = cv2.copyMakeBorder(img, top, bottom, left, right, cv2.BORDER_CONSTANT, value=color)  # pad
    return img, ratio, (dw, dh)

# Define scale_coords and clip_coords
def scale_coords(img1_shape, coords, img0_shape, ratio_pad=None):
    gain = min(img0_shape[0] / img1_shape[0], img0_shape[1] / img1_shape[1])
    pad = (img0_shape[1] - img1_shape[1] * gain) / 2, (img0_shape[0] - img1_shape[0] * gain) / 2
    coords[:, [0, 2]] -= pad[0]
    coords[:, [1, 3]] -= pad[1]
    coords[:, :4] /= gain
    clip_coords(coords, img0_shape)
    return coords

def clip_coords(coords, shape):
    coords[:, 0] = np.clip(coords[:, 0], 0, shape[1])
    coords[:, 1] = np.clip(coords[:, 1], 0, shape[0])
    coords[:, 2] = np.clip(coords[:, 2], 0, shape[1])
    coords[:, 3] = np.clip(coords[:, 3], 0, shape[0])

def plot_one_box(xyxy, img, color=None, label=None, line_thickness=3):
    """Plots a rectangle on an image with optional label."""
    xyxy = list(map(int, xyxy))
    cv2.rectangle(img, (xyxy[0], xyxy[1]), (xyxy[2], xyxy[3]), color or (255, 0, 0), line_thickness)

    if label:
        font_scale = line_thickness / 3
        font_thickness = max(line_thickness - 1, 1)
        text_size = cv2.getTextSize(label, 0, font_scale, font_thickness)[0]
        cv2.rectangle(img, xyxy[0], xyxy[1] - text_size[1] - 3, xyxy[0] + text_size[0], xyxy[1], color or (255, 0, 0), -1)  # Background rectangle
        cv2.putText(img, label, (xyxy[0], xyxy[1] - 2), 0, font_scale, [255, 255, 255], font_thickness, lineType=cv2.LINE_AA)

# Load model
weights = 'runs/train/exp3/weights/best.pt'  # Path to your trained model
device = select_device('')  # Use GPU if available
model = attempt_load(weights, device=device)  # Corrected line
model.eval()

# Initialize webcam
cap = cv2.VideoCapture(0)  # 0 for default webcam, change if needed

while cap.isOpened():
    ret, frame = cap.read()  # Capture frame-by-frame
    if not ret:
        print("Failed to grab frame")
        break

    # Preprocess the frame for YOLOv5
    img, _, _ = letterbox(frame, 640, stride=32)  # Modify to match function signature
    img = img[:, :, ::-1].transpose(2, 0, 1)  # BGR to RGB, to 3x416x416
    img = np.ascontiguousarray(img)

    img = torch.from_numpy(img).to(device)
    img = img.float()  # uint8 to fp16/32
    img /= 255.0  # Normalize image to [0, 1]
    if img.ndimension() == 3:
        img = img.unsqueeze(0)

    # Inference
    with torch.no_grad():
        pred = model(img, augment=False)[0]

    # Apply NMS (Non-Maximum Suppression)
    pred = non_max_suppression(pred, 0.25, 0.45)

    # Process detections
    for i, det in enumerate(pred):  # detections per image
        if len(det):
            # Rescale boxes from img_size to frame size
            det[:, :4] = scale_coords(img.shape[2:], det[:, :4], frame.shape).round()

            # Draw the bounding boxes and labels
            for *xyxy, conf, cls in reversed(det):
                label = f'{model.names[int(cls)]} {conf:.2f}'
                plot_one_box(xyxy, frame, label=label, color=[0, 255, 0], line_thickness=2)

    # Display the frame with detections
    cv2.imshow('Webcam YOLOv5', frame)

    if cv2.waitKey(1) == ord('q'):  # Press 'q' to quit
        break

cap.release()
cv2.destroyAllWindows()
```
